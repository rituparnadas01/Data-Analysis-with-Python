# Matplotlib

Matplotlib is a comprehensive library for creating various types of visualizations in Python. It offers extensive customization options and integrates well with other data science libraries like NumPy and Pandas.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Line Plots
Matplotlib's pyplot module provides a simple interface for creating various types of plots. Let's start with a basic line plot:

```python
# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

# Customizing Plots
Matplotlib offers extensive customization options for plots:

```python
# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a customized plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'b-', linewidth=2, label='sin(x)')
plt.plot(x, y2, 'r--', linewidth=2, label='cos(x)')
plt.title('Sine and Cosine Waves', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=5, color='k', linestyle='-', alpha=0.3)
plt.show()
```

# Different Plot Types
Matplotlib supports many types of plots for different data visualization needs:

```python
# Scatter Plot
plt.figure(figsize=(12, 10))

# Subplot 1: Scatter Plot
plt.subplot(2, 2, 1)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7)
plt.title('Scatter Plot')

# Subplot 2: Bar Chart
plt.subplot(2, 2, 2)
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 15]
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart')

# Subplot 3: Histogram
plt.subplot(2, 2, 3)
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title('Histogram')

# Subplot 4: Pie Chart
plt.subplot(2, 2, 4)
labels = ['Group A', 'Group B', 'Group C', 'Group D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```

# Subplots and Figure Layout
Matplotlib provides different ways to create multiple plots in a single figure:

```python
# Create a figure and a grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot on each subplot
x = np.linspace(0, 10, 100)

# Plot 1: Linear
axes[0, 0].plot(x, x, 'b-')
axes[0, 0].set_title('Linear')

# Plot 2: Quadratic
axes[0, 1].plot(x, x**2, 'r-')
axes[0, 1].set_title('Quadratic')

# Plot 3: Cubic
axes[1, 0].plot(x, x**3, 'g-')
axes[1, 0].set_title('Cubic')

# Plot 4: Square Root
axes[1, 1].plot(x, np.sqrt(x), 'y-')
axes[1, 1].set_title('Square Root')

# Adjust layout
plt.tight_layout()
plt.show()
```

# 3D Plotting
Matplotlib can create 3D visualizations with the mplot3d toolkit:

```python
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and colorbar
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

plt.show()
```

# Time Series Visualization
Matplotlib works well with time series data:

```python
# Create sample time series data
dates = pd.date_range('2024-01-01', periods=100)
values = np.cumsum(np.random.randn(100))

# Create time series plot
plt.figure(figsize=(12, 6))
plt.plot(dates, values, 'b-')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()

# Format the x-axis to show dates nicely
plt.gcf().autofmt_xdate()
plt.show()
```

# Data Visualization for Data Science
Let's demonstrate how Matplotlib can be used in a real data science workflow:

```python
# Create a sample dataset
data = {
    'Group': ['A', 'B', 'C', 'D', 'E'],
    'Value1': [10, 25, 15, 30, 20],
    'Value2': [8, 28, 12, 25, 22]
}
df = pd.DataFrame(data)

# Create visualizations
plt.figure(figsize=(15, 10))

# Subplot 1: Bar chart comparing groups
plt.subplot(2, 2, 1)
plt.bar(df['Group'], df['Value1'], color='skyblue', label='Value1')
plt.bar(df['Group'], df['Value2'], color='salmon', label='Value2', alpha=0.7, bottom=df['Value1'])
plt.title('Stacked Bar Chart by Group')
plt.xlabel('Group')
plt.ylabel('Values')
plt.legend()

# Subplot 2: Scatter plot with trend line
plt.subplot(2, 2, 2)
plt.scatter(df['Value1'], df['Value2'], s=100, c='purple', alpha=0.7)
# Add trend line
z = np.polyfit(df['Value1'], df['Value2'], 1)
p = np.poly1d(z)
plt.plot(df['Value1'], p(df['Value1']), 'r--', alpha=0.8)
plt.title('Relationship Between Value1 and Value2')
plt.xlabel('Value1')
plt.ylabel('Value2')

# Subplot 3: Box plot
plt.subplot(2, 2, 3)
plt.boxplot([df['Value1'], df['Value2']], labels=['Value1', 'Value2'])
plt.title('Distribution of Values')
plt.ylabel('Value')

# Subplot 4: Pie chart of Value1
plt.subplot(2, 2, 4)
plt.pie(df['Value1'], labels=df['Group'], autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Value1 by Group')

plt.tight_layout()
plt.show()
```

# Saving Figures
Matplotlib allows you to save your visualizations in various formats:

```python
# Create a simple plot
plt.figure(figsize=(10, 6))
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, 'g-', linewidth=2)
plt.title('Sine Wave')
plt.grid(True)

# Save the figure in different formats
plt.savefig('sine_wave.png', dpi=300, bbox_inches='tight')
plt.savefig('sine_wave.pdf', bbox_inches='tight')
plt.savefig('sine_wave.svg', bbox_inches='tight')

plt.show()
```
